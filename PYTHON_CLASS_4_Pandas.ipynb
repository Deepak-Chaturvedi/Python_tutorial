```python
# Group by
data3 = data2.iloc[10:20,[0,4,7,3,5,8,2,1]]
data3
```

```python
data3.groupby(['Team','Position']).get_group(('Boston Celtics', 'C'))
```

```python
#Find mean salary by groups 
print(data3.groupby(['Team','Position']).mean()[['Salary','Number']])
```

```python
data3.groupby('Position').groups
```

```python
grp = data3.groupby('Position')
for i,group in grp:
  print(i)
  print(group)
  print('\n\n\n')
```
